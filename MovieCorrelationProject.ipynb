```python
# import libraries 
import pandas as pd 
import numpy as np
import seaborn as sns 
import matplotlib 
import matplotlib.pyplot as plt 
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize']= (12,8)# to adjust configulation of plots to be created 

# read in the data 
df=pd.read_csv(r'C:\Users\LENOVO\Downloads\movies.csv')
```

```python
# checking missing data 
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{}-{}%'.format(col, pct_missing))
```

```python
# scatter plot between budget and gross
plt.scatter(x=df['budget'], y= df['gross'])
plt.title('Budget vs Gross Earnings')
plt.xlabel('Gross Earning')
plt.ylabel('Movie Budget')
plt.show()
```

```python
# using seaborn
sns.regplot(x= 'budget', y= 'gross', data= df, scatter_kws={'color': 'red'}, line_kws={'color': 'blue'})
```

```python
# correlations 
df.corr(method='kendall') # pearson, kendall ,spearman
```

```python
correlation_matrix = df.corr(method='spearman')
sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation Matrix for Numeric variables')
plt.xlabel(' Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
df_numerized = df
for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes

df_numerized.head()      
        
```

```python
correlation_matrix = df_numerized.corr(method='spearman')
sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation Matrix for Numeric variables')
plt.xlabel(' Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
df_numerized.corr()
```

```python
## unstacking 
correlation_mat = df_numerized.corr()
correlation_pairs = correlation_mat.unstack()
correlation_pairs
```

```python
sorted_pairs = correlation_pairs.sort_values()
sorted_pairs
```

```python
high_corr = sorted_pairs[(sorted_pairs)>0.5]
high_corr
```
